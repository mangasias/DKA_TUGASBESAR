```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
file_path = "2023-2024 NBA Player Stats - Playoffs.xlsx"
nba_df = pd.read_excel(file_path)

# Pilih fitur numerik dan label
features = ['GP', '3P', '3PA', '3P%', 'eFG%', 'Label']
correlation_matrix = nba_df[features].corr()

# Ambil korelasi terhadap 'Label' saja
correlations_with_label = correlation_matrix['Label'].drop('Label')

# Visualisasi korelasi
plt.figure(figsize=(8, 5))
sns.barplot(x=correlations_with_label.values, y=correlations_with_label.index, palette="viridis")
plt.title("Korelasi antara Fitur dan Kelayakan (Label)")
plt.xlabel("Korelasi Pearson")
plt.ylabel("Fitur")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

nba_df = pd.read_excel("2023-2024 NBA Player Stats - Playoffs.xlsx")

features = ['GP', '3P', '3PA', '3P%', 'eFG%']

plt.figure(figsize=(15, 8))
for i, feature in enumerate(features):
    plt.subplot(2, 3, i+1)
    sns.histplot(nba_df[feature], kde=True, bins=15, color='skyblue')
    plt.title(f'Distribusi {feature}')
plt.tight_layout()
plt.show()
```
